The Chocolate Sales Dataset (2025) is a dataset of chocolate product sales across multiple countries in 2025. It includes information about brands, product types, sales channels, countries and revenue.
Source: https://www.kaggle.com/datasets/syedaeman2212/chocolate-sales-dataset-in-2025

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('chocolate_sales_2025_dataset.csv')
df.head()
```

```python
# Prints details such as number of rows and index range, column names and data types, non-null value counts per column, and 
# memory usage (optional deep analysis)
df.info()
```

```python
# Generates descriptive statistics for a DataFrame to show the central tendency, dispersion, and shape of a dataset's distribution
df.describe()
```

The histogram demonstrates how the total revenue is distributed for each sale ID.

```python
# Histogram of Revenue_USD
plt.figure(figsize=(8, 5))
plt.hist(df['Revenue_USD'], bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Revenue (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Revenue per Sale')
plt.tight_layout()
plt.show()
```
